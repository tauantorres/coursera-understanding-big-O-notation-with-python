```python
import random
import time

def company_1(num_authors):
    # Represents O(1)
    print("Paying royalties to {} authors".format(num_authors))
    time.sleep(1) # Always takes 1 second

def company_2(num_authors):
    # Represents O(N)
    for i in range(num_authors):
        amount = random.randint(100, 1000)
        print("Paying ${} to author {}".format(amount, i))
        time.sleep(0.5) # Takes longer with more authors

def company_3(num_authors):
    # Represents O(N^2)
    for i in range(1, num_authors + 1):
        for j in range(1, num_authors + 1):
            amount = random.randint(100, 1000)
            print("Paying ${} to author {} via employee {}".format(amount, i, j))
            time.sleep(0.1) # Much longer with more authors
```

```python
%time company_1(100)
```

```python
%time company_2(10)
```

```python
%time company_3(10)
```

```python
import matplotlib.pyplot as plt

x_values = range(1, 100)

y1_values = [1 for x in x_values]
y2_values = [x for x in x_values]

fig, ax1 = plt.subplots(figsize= (12, 6))

color = 'tab:red'

ax1.set_xlabel(xlabel='Number of Authors')
ax1.set_ylabel(ylabel='Runtime', color=color)

ax1.plot(x_values, y1_values, color=color, label="O(1)")
ax1.plot(x_values, y2_values, color=color, label="O(N)")
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylim(bottom=0, top=100)
ax1.legend(loc="upper left")

ax2 = ax1.twinx()

y3_values = [x**2 for x in x_values]

color = 'tab:blue'

ax2.set_ylabel(ylabel='Runtime', color=color)
ax2.plot(x_values, y3_values, color=color, label="O(N^2)")
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(bottom=0, top=10000)
ax2.legend(loc="upper right")

fig.tight_layout()


plt.title("Big-O Scaling Comparison:  Three Publishing Companies Paying Royalty")
plt.show()
```

```python
# Task 1: Benchmark the Existing Runtimes:

import time

def benchmark(func, num_authors, label):
    start_time = time.time()
    func(num_authors)
    elapsed_time = time.time() - start_time
    print(f"{label} ({num_authors} authors): {elapsed_time:.2f} sec")

# Benchmark each function with 10 and 100 authors
print("Benchmark with 10 authors")
benchmark(company_1, 10, "Company 1")
benchmark(company_2, 10, "Company 2")
benchmark(company_3, 10, "Company 3")

print("\nBenchmark with 20 authors")
benchmark(company_1, 20, "Company 1")
benchmark(company_2, 20, "Company 2")
benchmark(company_3, 20, "Company 3")
```

```python
# Task 2: Refactor company_3 to O(N)

def company_3_refactored(num_authors):
    # Represents O(N)
    for i in range(num_authors):
        amount = random.randint(100, 1000)
        print("Paying ${} to author {}".format(amount, i))
        time.sleep(0.1)  # Adjusted for a faster, linear runtime

```

```python
# Task 3: Re-run Benchmarks and Analyze Efficiency Gain
print("Benchmark with 10 authors (optimized)")
benchmark(company_3_refactored, 10, "Company 3 Optimized")

print("\nBenchmark with 100 authors (optimized)")
benchmark(company_3_refactored, 100, "Company 3 Optimized")
```

```python
# Task 4: Plot Big-O Scalability

y3_refactored_values = [x for x in x_values]  # This line is O(N)

fig, ax1 = plt.subplots(figsize=(10, 5))

# Plot the original lines
ax1.plot(x_values, y1_values, label="O(1) - Company 1")
ax1.plot(x_values, y2_values, label="O(N) - Company 2")
ax1.plot(x_values, y3_values, label="O(N^2) - Company 3 Original")

# Plot the optimized version of company_3
ax1.plot(x_values, y3_refactored_values, label="O(N) - Company 3 Optimized", linestyle="--")

ax1.set_xlabel('Number of Authors')
ax1.set_ylabel('Runtime')
ax1.set_ylim(bottom=0, top=1000)
ax1.legend()
plt.title("Big-O Scaling Comparison: Original vs Optimized")

plt.show()
```
